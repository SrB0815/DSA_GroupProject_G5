Imputing missing values

```python
from sklearn.impute import SimpleImputer

df = df.copy()  
imputer_used = {}

for col in df.select_dtypes(include='number'):
    if col == 'Year':
        continue

    if abs(df[col].skew()) > 0.5:
        strategy = 'median'
    else:
        strategy = 'mean'

    imputer = SimpleImputer(strategy=strategy)
    df.loc[:, col] = imputer.fit_transform(df[[col]])

    imputer_used[col] = strategy
```

Feature Engineering

Population Density Category

```python
bins = [100, 500, 2000, 10000, 25000]
labels = ['Moderate', 'High', 'Very High', 'Extreme']

df['Population_Density_Category'] = pd.cut(
    df['Population_Density_Per_SqKm'],
    bins=bins,
    labels=labels
)
```

AQI category

```python
def aqi_category(AQI):
    if AQI<=50:
        return 'Good'
    elif AQI <=100:
        return 'Moderate'
    elif AQI <=150:
        return 'Sensitive'
    elif AQI <=200:
            return 'Unhealthy'
    elif AQI <=300:
         return 'Very Unhealthy'
    else:
         return 'Hazardous'

df['AQI_category']=df['AQI'].apply(aqi_category)
```

Categorizing Decades

```python
df['Decade'] = (df['Year'] // 10) * 10
df['Decade'] = df['Decade'].astype(str) + 's'
```

calculating CO2 per Population

```python
df['CO2_per_Population'] = df['CO2_Emissions_MT'] / (df['Population_Density_Per_SqKm'] + 1)
```

Calculating PM Ratio

```python
df['PM_Ratio'] = df['PM2.5'] / df['PM10']
```

Calculating Net Forest Change

```python
df['Net_Forest_Change_%'] = df['Afforestation_Rate_%'] - df['Deforestation_Rate_%']
```

Impact on Health due to Air Quality

```python
df['Pollution_Health_Impact'] = df['AQI'] / (df['Avg_Life_Expectancy_Index'] + 1)
```

Handling Ouliers

```python
for col in df.select_dtypes(include='number'):
    if col == 'Year':
        continue
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
def clean_data(df):
    df = df.copy() 

    df = df[(df['AQI'].between(0, 600))]
    df = df[(df['PM2.5'].between(0, 600))]
    df = df[(df['PM10'].between(0, 800))]
    
    df = df[(df['Deforestation_Rate_%'].between(0, 100))]
    df = df[(df['Industries_Increase_%'].between(0, 50))]
    df = df[(df['Vehicles_Increase_%'].between(0, 50))]
    
    df = df[df['CO2_Emissions_MT'] >= 0]
    
    df = df[(df['Green_Space_Ratio_%'].between(0, 100))]
    
    return df
```
